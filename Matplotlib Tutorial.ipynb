# creating subplots 

```python
plt.subplot(2,2,1)
plt.plot(a,b,c="r")
plt.subplot(2,2,2)
plt.plot(a,b,c="b")
plt.subplot(2,2,3)
plt.plot(a,b,c="y")
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,y)
y = np.cos(x)
plt.subplot(2,1,2)
plt.plot(x,y)
```
